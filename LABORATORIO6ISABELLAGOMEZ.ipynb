```python
import pandas as pd

salud = {
    'Nombre': ['Carlos', 'Ana', 'Luis', 'Pedro', 'María', None],
    'Edad': [34, 29, None, 45, 25, 33],
    'Fecha Ingreso': ['2021-01-15', '2020/03/12', '2022-07-01', '2021/12/01', '2021-05-20', '2020-03-12'],
    'Resultado Prueba': ['Positivo', 'Negativo', 'Positivo', 'Positivo', 'Negativo', 'Negativo']
}

df_salud = pd.DataFrame(salud)

#rellenar nombres e ingresos faltantes
df_salud['Nombre'] = df_salud['Nombre'].fillna('Desconocido')
df_salud['Edad'] = df_salud['Edad'].fillna(df_salud['Edad'].median())

#estandarizar fecha
df_salud['Fecha Ingreso'] = pd.to_datetime(df_salud['Fecha Ingreso'], format='mixed').dt.strftime('%Y-%m-%d')

print(df_salud)
```

```python
import pandas as pd

ecommerce = {
    'Cliente ID': [1, 2, 3, 4, 5],
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Cantidad': [10, 20, None, 5, 15],
    'Precio': [100, '200$', 300, 'cuatrocientos', 500],
    'Fecha Transacción': ['2023-01-15', '2023/02/20', '2023-03-25', '2023/04/30', '2023-05-10']
}

df_eco = pd.DataFrame(ecommerce)

#limpiar precios
df_eco['Precio'] = df_eco['Precio'].replace({'cuatrocientos': 400})
df_eco['Precio'] = df_eco['Precio'].replace(r'\$', '', regex=True).astype(float)

#rellenar cantidades faltantes
df_eco['Cantidad'] = df_eco['Cantidad'].fillna(0)

#formatear fechas
df_eco['Fecha Transacción'] = pd.to_datetime(df_eco['Fecha Transacción'], format='mixed').dt.date

print(df_eco)
```

```python
import pandas as pd

#cargar datos
ventas = {
    'Nombre': ['Juan', 'Maria', 'Pedro', 'Ana', 'Luisa', 'Carlos', 'Juan'],
    'Edad': [25, 30, None, 29, 28, 200, 25],
    'Email': ['juan@mail.com', 'maria@mail.com', 'pedro@mail.com', 'ana@mail.com', 'luisa@mail.com', 'carlos@mail.com', 'juan@mail.com'],
    'Pais': ['colombia', 'Colombia', 'Mexico', 'Argentina', 'Colombia', 'Brazil', 'colombia'],
    'Genero': ['M', 'F', 'H', 'F', 'Femenino', 'M', 'M'],
    'Visitas': [5, 7, 6, 4, 8, 2, 5]
}

df_ventas = pd.DataFrame(ventas)

#eliminar duplicados
df_ventas = df_ventas.drop_duplicates()

#corregir edad inválida
#reemplazamos por la media de las edades válidas
media_edad = df_ventas[df_ventas['Edad'] < 100]['Edad'].mean()
df_ventas.loc[df_ventas['Edad'] > 100, 'Edad'] = media_edad
df_ventas['Edad'] = df_ventas['Edad'].fillna(media_edad)

#estandarizar generos
mapeo_genero = {'M': 'Masculino', 'H': 'Masculino', 'F': 'Femenino', 'Femenino': 'Femenino'}
df_ventas['Genero'] = df_ventas['Genero'].map(mapeo_genero)

#estandarizar paises
df_ventas['Pais'] = df_ventas['Pais'].str.capitalize()

print(df_ventas)
```
